## Scaling of Numeric Data

```python
import timeit
import numpy as np
from sklearn import datasets, linear_model 

X, y = datasets.load_diabetes(return_X_y=True)
raw = X[:, None, 2]
```

```python
# Scaling rules
max_raw = max(raw)
min_raw = min(raw)
scaled = (2*raw - max_raw - min_raw)/(max_raw - min_raw)

fig, axs = plt.subplots(2, 1, sharex=True )

axs[0].hist(raw)
axs[1].hist(scaled)
```

```python
# Training models
def train_raw():
  linear_model.LinearRegression().fit(raw, y)

def train_scaled():
  linear_model.LinearRegression().fit(scaled, y)
```

```python
raw_time = timeit.timeit(train_raw, number=100)
scaled_time = timeit.timeit(train_scaled, number=100)
print('train raw: {}'.format(raw_time))
print('train scaled: {}'.format(scaled_time))
```

```python
z_score=(raw - np.mean(raw))/np.std(raw)

fig, axs = plt.subplots(2, 1, sharex=True )

axs[0].hist(raw)
axs[1].hist(z_score)
```

```python
def train_z_score():
  linear_model.LinearRegression().fit(z_score, y)
```

```python
raw_time = timeit.timeit(train_raw, number=100)
scaled_time = timeit.timeit(train_scaled, number=100)
z_score_time = timeit.timeit(train_z_score, number=100)

print('train raw: {}'.format(raw_time))
print('train scaled: {}'.format(scaled_time))
print('train z-score: {}'.format(z_score_time))
```

## Dimensionality Reduction with PCA

```python
from sklearn.preprocessing import StandardScaler

iris = sns.load_dataset('iris')

scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
)
covariance_matrix = np.cov(scaled.T)
covariance_matrix
```

```python
sns.pairplot(iris)
```

```python
sns.jointplot(x=iris['petal_length'], y=iris['petal_width'])
sns.jointplot(x=scaled[:, 2], y=scaled[:, 3])
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
variance_explained = []
for i in eigen_values:
  variance_explained.append((i/sum(eigen_values))*100)

print(variance_explained)
```

```python
from sklearn.decomposition import PCA 

pca = PCA(n_components=2)
pca.fit(scaled)
```

```python
pca.explained_variance_ratio_
```

```python
reduce_scaled = pca.transform(scaled)
```

```python
iris['pca_1'] = scaled[:,0]
iris['pca_2'] = scaled[:,1]
iris
```

```python
sns.jointplot(iris['pca_1'], iris['pca_2'], hue=iris['species'])
```
